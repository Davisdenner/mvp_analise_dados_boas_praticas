# Descrição do Problema

Este conjunto de dados explora a relação entre desempenho acadêmico e sucesso na carreira. Ele inclui 5.000 registros com informações sobre a formação educacional dos estudantes, habilidades e resultados profissionais. O arquivo, chamado [education_career_success](https://www.kaggle.com/datasets/adilshamim8/education-and-career-success), foi 100% extraído do Kaggle. O principal objetivo deste conjunto de dados é identificar os fatores que mais influenciam os salários e compreender o papel do networking e dos estágios no crescimento profissional, utilizando a linguagem de programação Python como base para a análise.


## Hipóteses do Problema

Explorar a relação entre o desempenho acadêmico e o sucesso profissional de estudantes que moram nos Estados Unidos (EUA), com base nas seguintes perguntas:


- Quais fatores mais contribuem para um crescimento rápido na carreira (years_to_promotion)?

- Estudantes de certas áreas (field_of_study) tem mais oportunidades de emprego do que outras?

- O número de estágios concluídos (interships_completed) tem impacto no salário inicial?



## Tipo de Problema

Este é um problema de aprendizado não supervisionado. Com base em características como universidade, área de estudo, habilidades, estágios e networking, o objetivo futuramente é identificar padrões ou agrupar estudantes com perfis semelhantes, sem uma variável-alvo predefinida.


## Atributos do Dataset
O dataset **Education & Career Success** reúne informações abrangentes sobre os estudantes, incluindo desempenho acadêmico, competências, atividades extracurriculares e os respectivos resultados profissionais. A seguir, são apresentadas as descrições detalhadas das colunas, facilitando o entendimento e a análise do conjunto de dados.



###**Informações do estudante**

- **Student_ID**: Identificador único para cada estudante;

- **Age**: Idade do estudante (18-30 anos);

- **Gender**: Gênero ( Masculino, Feminino ou Outro);


###**Desempenho acadêmico**

- **High_School_GPA**: Média escolar do ensino médio (escala de 2.0 - 4.0);

- **SAT_Score**: Pontuação no teste padronizado (900 - 1600);

- **University_Ranking**: Ranking da universidade frequentada (1-1000);

- **University_GPA**: Média acadêmica na universidade (escala de 2.0 - 4.0);

- **Field_of_Study**: Área de estudo (Ciência da Computação, Medicina, Administração/Negócios, Artes, Direito, Engenharia, Matemática);

###**Habilidades e atividades extracurriculares**

- **Internships_Completed**:  Número de estágios concluídos (0-4);

- **Projects_Completed**:  Número de projetos acadêmicos ou pessoais concluídos (0-9);
- **Certifications**:  Número de certificações adicionais obtidas (0-5);

- **Soft_Skills_Score**:  Avaliação das habilidades interpessoais (1-10);

- **Networking_Score**: Pontuação baseada em conexões e networking profissional (1-10);


###**Resultados profissionais**

- Job_Offers: Número de ofertas de emprego recebidas após a graduação (0-5);

- Starting_Salary: Salário inicial (anual) no primeiro emprego em dólares americanos (25.000 - 101.000);

- Career_Satisfaction: Nível de satisfação com a carreira (1-10);

- Years_to_Promotion: Tempo necessário para receber a primeira promoção (1-5 anos);

- Current_Job_Level: Nível de carreira atual (Entry, Mid, Senior, Executive);

- Work_Life_Balance: Avaliação do equilíbrio entre trabalho e vida pessoal (1-10);

- Entrepreneurship: Indica se a pessoa iniciou um negócio próprio (Sim/Não).

### Média

A média é uma medida de tendência central que representa o valor típico ou o ponto de equilíbrio de um conjunto de dados. É calculada somando-se todos os valores e dividindo-se pelo número total de observações. É sensível a valores extremos (outliers).

```python
# média dos atributos numéricos do dataset
df.describe().loc['mean'].round(2)
```

### Média do salário inicial por ranking de universidade

Um gráfico interessante para ilustrar como a medida de média foi utilizada neste dataset é aquele que mostra a **média do salário inicial por ranking de universidade.** Esse gráfico pode ser utilizado para responder à pergunta: **“Estudantes de universidades mais bem classificadas ganham salários iniciais mais altos?”**



```python
# calculando a média do salário inicial por ranking de universidade
avg_salary_by_ranking = df.groupby('University_Ranking')['Starting_Salary'].mean().reset_index()

# arredondando o salário médio para 2 casas decimais
avg_salary_by_ranking['Starting_Salary'] = avg_salary_by_ranking['Starting_Salary'].round(2)

# limitaando aos 10 primeiros rankings
avg_salary_by_ranking_top10 = avg_salary_by_ranking.head(10).sort_values(by='University_Ranking', ascending=True)

# criando o gráfico de barras usando plotly.express
fig = px.bar(avg_salary_by_ranking_top10,
             x='University_Ranking',
             y='Starting_Salary',
             color='Starting_Salary',
             text = 'Starting_Salary',
             color_continuous_scale='Viridis',
             title='Salário Médio Inicial por Ranking de Universidade (Top 10)')

# centralizando o título do gráfico
fig.update_layout(title_x=0.5)


fig.show()
```

Conforme os dados apresentados, é possível perceber que **nem sempre os estudantes de universidades mais bem classificadas possuem os salários iniciais mais altos**, uma vez que os alunos com a maior média de salário inicial frequentaram universidades que ocupam a **sétima e oitava posição no ranking**, e não as primeiras.

### Desvio Padrão

O desvio padrão é uma medida de dispersão que quantifica a quantidade de variação ou dispersão de um conjunto de valores. Um desvio padrão baixo indica que os pontos de dados tendem a estar próximos da média do conjunto, enquanto um desvio padrão alto indica que os pontos de dados estão espalhados por uma faixa maior de valores. Ele é a raiz quadrada da variância.

```python
# desvio padrão dos atributos numéricos do dataset
df.std(numeric_only=True).round(2)
```

**Por que alguns desvios padrões estão tão altos?**

Variáveis como Starting_Salary têm valores na casa dos milhares. Se o salário médio for, por exemplo, 60.000, então um desvio padrão de 14.494,96 não é absurdo, significa que há uma variação natural em torno desse valor médio.

A mesma lógica vale para **SAT_Score** e **University_Ranking**, que também têm escalas naturalmente maiores.

## Por que a correlação é tão baixa?

Investigando a base de dados, foi possível chegar as seguintes conclusões.

**1. Variáveis discretas em grande parte:**

- Variáveis como "Certifications", "Job_Offers", "Internships_Completed", "Projects_Completed" e "Years_to_Promotion" são números pequenos e discretos (ex.: 0 a 5). Isso naturalmente tende a gerar correlações fracas, especialmente quando há poucos níveis possíveis.

**2. Baixa relação linear:**

- As relações entre os fatores provavelmente não são lineares. Por exemplo, o fato de alguém ter 3 certificações não necessariamente impacta linearmente o salário ou as promoções.

**3. Dataset balanceado propositalmente:**

- Como os valores são bem distribuídos, é possível que o dataset tenha sido gerado ou tratado para evitar vieses, o que dilui as correlações.

**4. Fatores qualitativos mascarados em números:**

- Métricas como Soft_Skills_Score ou Work_Life_Balance são subjetivas e podem não ter uma ligação direta e objetiva com salários, promoções ou satisfação.

**5. Sistemas Complexos:**

- Na vida real, crescimento de carreira não depende de apenas um fator isolado, mas de interações complexas entre vários fatores, que uma matriz de correlação simples não captura.

# Pré-Processamento de Dados

O pré-processamento de dados é uma etapa crucial para preparar os dados para a modelagem, garantindo que estejam no formato adequado e otimizados para o desempenho dos algoritmos.

Como se trata de um problema de aprendizado não supervisionado, não haverá distinção entre features (X) e target (y); todas as colunas relevantes serão consideradas como features.

A coluna Student_ID não será útil para identificar padrões ou agrupar perfis, portanto, será removida do conjunto de dados para a análise não supervisionada.

```python
#criando um novo DataFrame X para o pré-processamento e modelagem não supervisionada
X = df.copy()
```

```python
#removendo colunas que não serão usadas como features para o aprendizado não supervisionado
if 'Student_ID' in X.columns:
    X = X.drop('Student_ID', axis=1)
```

```python
#Visualizando as primeiras linhas do DataFrame preparado para análise não supervisionada
print("DataFrame X para análise não supervisionada:")
X.head()
```

## Verficiando a existencia de números nulos

```python
# verificando a presença de valores nulos no dataset original
print("Valores nulos no dataset Education & Career Success:")
df.isnull().sum()
```

 ## Identificando colunas numéricas e categóricas

```python
# identificando colunas numéricas e categóricas

# selecionando as colunas numéricas (int64 e float64)
colunas_numericas = X.select_dtypes(include=['int64', 'float64']).columns

# selecionando as colunas categóricas (object ou category)
# exclude para garantir que pegamos os tipos categóricos
colunas_categoricas = X.select_dtypes(exclude=['int64', 'float64']).columns

print(f"Colunas Numéricas: {list(colunas_numericas)}")
print(f"Colunas Categóricas: {list(colunas_categoricas)}")
```

## Aplicando StandardScaler nas colunas numéricas

```python
#aplicando StandardScaler nas colunas numéricas

# criando uma instância do StandardScaler
scaler = StandardScaler()

# aplicando o scaler apenas às colunas numéricas do DataFrame X
# O .fit_transform() calcula a média e o desvio padrão e depois transforma os dados
X_numericas_escaladas = scaler.fit_transform(X[colunas_numericas])

# O resultado de fit_transform é um array NumPy.
# Para facilitar a combinação futura, é útil convertê-lo de volta para um DataFrame,
# mantendo os nomes das colunas originais.
X_numericas_escaladas_df = pd.DataFrame(X_numericas_escaladas, columns=colunas_numericas, index=X.index)

# visualizando as primeiras linhas do DataFrame numérico escalonado
print("\nPrimeiras linhas das colunas numéricas escalonadas:")
X_numericas_escaladas_df.head()
```

## Aplicando OneHotEncoder nas colunas categóricas

```python
# Aplicando OneHotEncoder nas colunas categóricas

# Criando uma instância do OneHotEncoder
# handle_unknown='ignore' é útil para evitar erros se houver valores inesperados nos dados futuros
# sparse_output=False retorna um array denso, mais fácil de trabalhar inicialmente com pandas
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

# Aplicando o encoder apenas às colunas categóricas do DataFrame X
# O .fit_transform() aprende as categorias únicas e as codifica
X_categoricas_codificadas = encoder.fit_transform(X[colunas_categoricas])

# o resultado de fit_transform é um array NumPy.
# para facilitar a combinação futura, é útil convertê-lo de volta para um DataFrame.
# obtendo os nomes das novas colunas criadas pelo encoder.
nomes_novas_colunas_categoricas = encoder.get_feature_names_out(colunas_categoricas)
X_categoricas_codificadas_df = pd.DataFrame(X_categoricas_codificadas, columns=nomes_novas_colunas_categoricas, index=X.index)

# visualizando as primeiras linhas do DataFrame categórico codificado
print("\nPrimeiras linhas das colunas categóricas codificadas (One-Hot):")
X_categoricas_codificadas_df.head()
```

## Combinando as colunas processadas em um único DataFrame/Array para o algoritmo de clustering

```python
# aplicando StandardScaler nas colunas numéricas

# criando uma instância do StandardScaler
scaler = StandardScaler()

# Aplicando o scaler apenas às colunas numéricas do DataFrame X
# O .fit_transform() calcula a média e o desvio padrão e depois transforma os dados
X_numericas_escaladas = scaler.fit_transform(X[colunas_numericas])

# O resultado de fit_transform é um array NumPy.
# Para facilitar a combinação futura, é útil convertê-lo de volta para um DataFrame,
# mantendo os nomes das colunas originais.
X_numericas_escaladas_df = pd.DataFrame(X_numericas_escaladas, columns=colunas_numericas, index=X.index)

# visualizando as primeiras linhas do DataFrame numérico escalonado
print("\nPrimeiras linhas das colunas numéricas escalonadas:")
X_numericas_escaladas_df.head()
```

PCA - Entendendo a estrutura linear dos dados.

Redução de Dimensionalidade: Após a aplicação do One-Hot Encoding, o conjunto de dados passou a ter 10 features numéricas e 16 features categóricas, totalizando 26 features.

Isso pode dificultar a visualização dos clusters e até impactar o desempenho de alguns algoritmos. Técnicas como PCA (Análise de Componentes Principais) podem ser utilizadas para reduzir a dimensionalidade, preservando a maior parte da variância dos dados.

```python
# Combinando as colunas processadas em um único DataFrame/Array para o algoritmo de clustering
# Concatenando os DataFrames das colunas numéricas escalonadas e categóricas codificadas
X_processado = pd.concat([X_numericas_escaladas_df, X_categoricas_codificadas_df], axis=1)

print("\nPrimeiras linhas do DataFrame processado (escalonado e codificado):")
print(X_processado.head())

# O número de componentes pode ser chosen com base na variância explicada ou para visualização (n_components=2 ou 3)
# começando com 2 componentes para visualização
pca = PCA(n_components=2)

# Aplicando o PCA ao seu DataFrame processado (X_processado)
X_pca = pca.fit_transform(X_processado)

# X_pca agora é um array NumPy com as duas principais componentes principais
print(f"\nFormato dos dados após PCA: {X_pca.shape}")

#visualizando a variância explicada pelos componentes
print(f"Variância explicada pelas 2 primeiras componentes: {pca.explained_variance_ratio_.sum():.4f}")

# Agora pode ser possível usar o X_pca para visualizações ou como entrada para algoritmos de clustering.
```

12,47% é uma variância explicada baixa para as duas primeiras componentes, o que significa que, juntas, elas conseguem capturar apenas cerca de 12,47% da variabilidade total presente no conjunto de dados processado. Isso indica que um PCA em 2D não representa bem os dados. Portanto, é mais adequado utilizar um PCA com um número maior de componentes para alimentar os algoritmos de clustering.”

```python
# Usando PCA com um número maior de componentes para clustering
# Substituindo 15 pelo número de componentes para aumentar tentar aumentar a porcentagem da variância
n_components_for_clustering = 15

pca_for_clustering = PCA(n_components=n_components_for_clustering)
X_pca_for_clustering = pca_for_clustering.fit_transform(X_processado)

print(f"\nFormato dos dados após PCA para clustering ({n_components_for_clustering} componentes): {X_pca_for_clustering.shape}")
print(f"Variância explicada por {n_components_for_clustering} componentes: {pca_for_clustering.explained_variance_ratio_.sum():.4f}")
```

Foram selecionadas 15 componentes principais, que explicam mais de 86% da variância total dos dados originais, que possuíam 26 features. Com isso, foi possível reduzir a dimensionalidade de 26 para 15 variáveis, mantendo a maior parte da informação relevante do conjunto de dados.

Essa redução representa aproximadamente 42% a menos de variáveis, configurando um trade-off eficiente entre simplificação do modelo e preservação da variabilidade dos dados. Essa abordagem é especialmente vantajosa para algoritmos de aprendizado não supervisionado, como clustering, pois melhora o desempenho computacional e reduz o impacto de ruídos e redundâncias nas variáveis.

```python
#Gráfico da Variância Explicada Cumulativa
explained_variance_ratio = pca_for_clustering.explained_variance_ratio_
cumulative_explained_variance = explained_variance_ratio.cumsum()

plt.figure(figsize=(10, 6))
plt.plot(range(1, n_components_for_clustering + 1), explained_variance_ratio, marker='o', linestyle='--', label='Variância Explicada por Componente')
plt.plot(range(1, n_components_for_clustering + 1), cumulative_explained_variance, marker='o', linestyle='-', color='red', label='Variância Explicada Cumulativa')
plt.title('Variância Explicada pelas Componentes Principais')
plt.xlabel('Número da Componente Principal')
plt.ylabel('Proporção da Variância Explicada')
plt.xticks(range(1, n_components_for_clustering + 1))
plt.grid(True)
plt.legend()
plt.show()
```

**Análise do Gráfico de Variância Explicada**

O gráfico demonstra que as componentes principais apresentam variâncias muito próximas e relativamente baixas de forma individual, conforme evidenciado pela linha azul, que se mantém praticamente estável e em valores reduzidos.

Observa-se também que a variância explicada acumulada cresce de maneira quase linear, representada pela linha vermelha. Esse comportamento sugere que nenhuma variável isoladamente se destaca de forma significativa na explicação da variabilidade dos dados. Portanto, o conjunto de dados apresenta-se relativamente equilibrado, sem a presença de variáveis altamente dominantes.
